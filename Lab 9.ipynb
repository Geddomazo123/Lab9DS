## Normalización

```python
dfn=pd.melt(df,id_vars=['Año','Mes','Día','Fecha'],var_name='Combustible',value_name='Consumo')

# dfn.loc[dfn.Combustible.isin(['Asfalto','Diesel']),]
dfn
```

## Crecimiento Interanual

```python
dfn['Crecimiento Interanual']=np.nan
for i in combustibles:
    dfn.loc[dfn.Combustible==i,'Crecimiento Interanual'] = dfn.loc[dfn.Combustible==i,'Consumo'].pct_change(12)
    
```

```python
dfn.loc[dfn.Combustible.str.contains('Regular'),'Categoría']='Gasolina Regular'
dfn.loc[dfn.Combustible.str.contains('Súper'),'Categoría']='Gasolina Súper'
dfn.loc[dfn.Combustible.str.contains('Total'),'Categoría']='Gasolina Total'
dfn.loc[dfn.Combustible.str.contains('Diesel'),'Categoría']='Diesel'
dfn
```

```python
dfn.to_csv('DatosFinalesFinales.csv',index=False)
```
